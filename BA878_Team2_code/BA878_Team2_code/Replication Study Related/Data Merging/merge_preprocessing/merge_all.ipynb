## Data Input

```python
df_exercise = pd.read_csv('exercise_merged-2.csv')
df_exercise.head()
```

## merge all together

```python
df_merge = (
    df_basic_hormones_glucose
    .merge(df_sleep, on=["id", "day_in_study"], how="left", suffixes=("", "_slp"))
    .merge(df_hr, on=["id", "day_in_study"], how="left", suffixes=("", "_hr"))
    .merge(df_exercise, on=["id", "day_in_study"], how="left", suffixes=("", "_exe")))
```

```python
def check_unique(df, name):
    is_unique = not df.duplicated(["id", "day_in_study"]).any()
    dup_count = df.duplicated(["id", "day_in_study"], keep=False).sum()
    print(f"{name}: {'✅ Unique' if is_unique else '⚠️ Not Unique'}, it has {dup_count} duplicates")


check_unique(df_sleep, "df_sleep")
```
