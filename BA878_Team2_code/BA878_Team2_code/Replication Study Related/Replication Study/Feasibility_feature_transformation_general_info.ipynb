```python
df = df[df["study_interval"] == 2022]
df["age"] = df["study_interval"] - df["birth_year"]
df = df[["id", "day_in_study", "phase", "resting_heart_rate", 
         "steps", "calories", "minutesasleep",
        "flow_volume", "headaches", "cramps", "sorebreasts", 
         "fatigue", "moodswing", "indigestion", "bloating", "age"]]
# Need wrist temp, hrv_rmssd, respiratory rate, active minutes, BMI, VO2 max

df.head()
```

```python
# Create a "week" variable
df["week_in_study"] = (df["day_in_study"] / 7).astype(int)

# Count unique participants per week
participants_per_week = (
    df.groupby("week_in_study")["id"]
      .nunique()
      .reset_index(name = "n_active"))
participants_per_week["retention"] = (
    participants_per_week["n_active"] / participants_per_week["n_active"].iloc[0])

print(participants_per_week)
```

```python
duration_df = (
    df.groupby("id")
      .agg(
          first_day=("day_in_study", "min"),
          last_day=("day_in_study", "max")
      )
      .reset_index()
)

duration_df["duration_days"] = duration_df["last_day"] - duration_df["first_day"]

print(duration_df)
```

```python
df["fitbit_valid"] = df[["steps", "resting_heart_rate"]].notna().any(axis=1)
```

```python
questionnaire_cols = ["headaches", "cramps", "moodswing", "fatigue", "bloating"]
df["questionnaire_completed"] = df[questionnaire_cols].notna().any(axis=1)
```

```python
summary_df = (
    df.groupby("id")
      .agg(
          total_days=("day_in_study", "nunique"),
          fitbit_days=("fitbit_valid", "sum"),
          questionnaire_days=("questionnaire_completed", "sum")
      )
      .reset_index()
)

summary_df["pct_fitbit_valid"] = 100 * summary_df["fitbit_days"] / summary_df["total_days"]
summary_df["pct_questionnaire_completed"] = 100 * summary_df["questionnaire_days"] / summary_df["total_days"]

print(summary_df.head())
```

```python
total_days = df["day_in_study"].count() 
fitbit_valid_days = df["fitbit_valid"].sum()
questionnaire_valid_days = df["questionnaire_completed"].sum()

overall_fitbit_pct = 100 * fitbit_valid_days / total_days
overall_questionnaire_pct = 100 * questionnaire_valid_days / total_days

print(f"Overall % days with valid Fitbit data: {overall_fitbit_pct:.2f}%")
print(f"Overall % days with completed questionnaires: {overall_questionnaire_pct:.2f}%")
```

```python
import pandas as pd

# 0. Sort first
df = df.sort_values(["id", "day_in_study"]).reset_index(drop=True)

prev_phase = df.groupby("id")["phase"].shift(1)

is_first_row = df.groupby("id").cumcount() == 0
new_cycle_flag = (
    ((df["phase"] == "Follicular") & (prev_phase == "Luteal")) | is_first_row
).astype(int)

df["cycle"] = new_cycle_flag.groupby(df["id"]).cumsum()

wearable_cols = ["resting_heart_rate", "steps", "calories", "minutesasleep"]

df["has_phase"] = df["phase"].notna()
df["has_wearable"] = df[wearable_cols].notna().any(axis=1)
df["has_both"] = df["has_phase"] & df["has_wearable"]

cycle_summary = (
    df.groupby(["id", "cycle"])
      .agg(
          ovulatory=("phase", lambda s: (s == "Fertility").any()),
          partial_wearable_phase=("has_both", "any"),
      )
      .reset_index()
)

N = cycle_summary.query("ovulatory and partial_wearable_phase").shape[0]
print("N of ovulatory cycles with partial wearable + phase data:", N)
```

```python
df = df.sort_values(["id", "cycle", "day_in_study"])

cycle_len = (
    df.groupby(["id", "cycle"])["day_in_study"]
      .agg(lambda s: s.max() - s.min() + 1)
      .reset_index(name="cycle_length")
)
```

```python
df = df.merge(cycle_len, on=["id", "cycle"], how="left")
```

```python
target_phases = ['Luteal', 'Menstrual']
subset = df[df['phase'].isin(target_phases)]

total_reports = len(subset)

avg_reports = subset.groupby('id')['day_in_study'].count().mean()

print(f"Total reports during Luteal & Menses: {total_reports}")
print(f"Average reports per participant: {avg_reports:.2f}")
```
